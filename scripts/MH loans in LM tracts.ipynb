```python
import numpy as np
import pandas as pd
import geopandas as gp

import os
```

```python
bay_county_names = ["Alameda", "ContraCosta", "Sonoma", "Solano", "SanMateo", "SantaClara", "SanFrancisco", "Marin","Napa"] 
recovery_years = ["2013", "2014", "2015", "2016", "2017"]
data_path = '/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/data/parsed_data/'
```

```python
#get all the files for all bay area counties for years in the recovery period and concatenate them
files = [f for f in os.listdir(data_path) if f[-3:] == "csv" and f.split('_')[1] in recovery_years]
dfs = []
for file in files:
    df = pd.read_csv(data_path+file)
    dfs.append(df)

all_recovery = pd.concat(dfs, axis=0)
all_recovery
```

```python
#take the subset of the tracts that are CRA eligible and save them to a new dataframe
LMItracts = all_recovery[all_recovery["CRA Eligible"] == "eligible"]
```

```python
#crosswalk used to assign census tracts to places
crosswalk = pd.read_csv("/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/data/reference/geocorr2014.csv")
crosswalk["Place name"].replace(to_replace=' ', value='unincorporated non CDP', inplace=True)
crosswalk.head()
```

```python
recovery_places_LMI = pd.merge(crosswalk, LMItracts, how='right', left_on='Tract', right_on='Tract') #code by place
#because this dataframe only has CRA eligible tracts, total loans is the sum of the loans to high income borrowers
#and low income borrowers in low income tracts only
recovery_places_LMI["Total Loans"] = recovery_places_LMI["# HI borrower, LI tract"]+ recovery_places_LMI["# LI borrower, LI tract"]

recovery_places_LMI.rename(columns={"# LI borrower, LI tract": "Loans to LM borrowers", 
                             "# HI borrower, LI tract": "Loans to MH borrowers"}, inplace=True)

recovery_places_LMI["CRA Eligible"] = recovery_places_LMI["CRA Eligible"].apply(lambda x: x.replace("not eligible", "0").replace("eligible", "1"))
recovery_places_LMI["CRA Eligible"].astype(int)
recovery_places_LMI.head(10)
```

```python
df_lol = []
for place, place_df in recovery_places_LMI.groupby("Place name"):
    pct_LCTs_HIB = place_df["Loans to MH borrowers"].sum()/place_df["Total Loans"].sum()
    
    row_list = [place, place_df["County name"].iloc[0], place_df["CRA Eligible"].iloc[0], place_df["Total Loans"].sum(), place_df["Loans to LM borrowers"].sum(),
                place_df["Loans to MH borrowers"].sum(), len(place_df["CRA Eligible"]), pct_LCTs_HIB]
    df_lol.append(row_list)
                
                
recovery_loans_byincome_place = pd.DataFrame(df_lol, columns=["Place Name", "County", "CRA Eligibility", "Total loans 2013-2017", "LM loans 2013-2017", 
    "MH loans 2013-2017", "LMI tracts", "pct loans to MH borrowers in LM tracts"])
recovery_loans_byincome_place.head()
#data.to_csv("/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/results/recovery_loans_byincomeandplace.csv", index=False)
```

```python
def make_place_name(old_place_name):
    new_name = old_place_name.replace(' city, CA', '')
    new_name = new_name.replace( ' CDP, CA', '')
    return new_name
```
