```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.base.model import GenericLikelihoodModel
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

#### 2013-2014 


```python
demog_with_cra_dummies1314 = pd.read_csv("/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/data/regression/demog_with_cra_dummies_201314.csv")
```

```python
X_vars_models_1314 = demog_with_cra_dummies1314.loc[:, ~demog_with_cra_dummies1314.columns.isin(['Geoid', "Geoid_x","Geoid_y", "_merge", 'Total Loans',
         'type', "CRA Eligible", 'Year', "Year_x", "Tract_y", '% HI borrower, LI tract',
       '# HI borrower, LI tract', '% HI borrower, HI tract',
       '# HI borrower, HI tract', '% LI borrower, LI tract',
       '# LI borrower, LI tract', '%LI borrower, HI tract',
       '# LI borrower, HI tract' ])]
Y_vars_models_1314 = demog_with_cra_dummies1314['Total Loans']
```

```python
X_1314_constant = sm.add_constant(X_vars_models_1314)[['Percent NH White alone',"Total number of housing units", '% Single Family',
    '% Owner Occupied','Median home value','Median income',
    'Percent of population 25 years and over with Bachelor\'s degree',
    '% below poverty level', "County_Alameda County", "County_Contra Costa County", "County_Marin County", 
          "County_San Francisco County", "County_San Mateo County", "County_Santa Clara County",
              "County_Solano County", "County_Sonoma County", "CRA"]]
m = sm.regression.linear_model.OLS(Y_vars_models_1314, X_1314_constant)
results_m = m.fit()
print(results_m.summary())
with open('/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/results/total_loan_regressions/OLStotalloans_201314.html', 'w') as outfile:
    outfile.write(results_m.summary().as_html())

```

```python
vif = pd.DataFrame()
no_counties = X_vars_models_1314.loc[:, ~X_vars_models_1314.columns.isin(["County_Alameda County", "County_Contra Costa County", "County_Marin County", 
          "County_San Francisco County", "County_San Mateo County", "County_Santa Clara County",
              "County_Solano County", "County_Sonoma County"])]
vif["VIF Factor new model 3.05"] = [variance_inflation_factor(no_counties.values, i) for i in range(no_counties.shape[1])]
vif["features"] = no_counties.columns
vif.round(2)
vif.to_csv("/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/results/collinearity/VIF_OLStotalloans_201314.csv")
vif
```
