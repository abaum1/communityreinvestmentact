```python
import numpy as np
import pandas as pd
import geopandas
import os
```

```python
bay_county_names = ["Alameda", "ContraCosta", "Sonoma", "Solano", "SanMateo", "SantaClara", "SanFrancisco", "Marin","Napa"] 
recovery_years = ["2013", "2014", "2015", "2016", "2017"]
data_path = '/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/data/parsed_data/'
```

```python
#get all the files for all bay area counties for years in the recovery period and concatenate them
files = [f for f in os.listdir(data_path) if f[-3:] == "csv" and f.split('_')[1] in recovery_years]
dfs = []
for file in files:
    df = pd.read_csv(data_path+file)
    dfs.append(df)

all_recovery = pd.concat(dfs, axis=0)
all_recovery.head()
```

```python
#take the subset of the tracts that are CRA eligible and save them to a new dataframe
LMtracts = all_recovery[all_recovery["CRA Eligible"] == "eligible"]
```

```python
#crosswalk used to assign census tracts to places
crosswalk = pd.read_csv("/Users/ameliabaum/Desktop/Amelia/CRA_Thesis/communityreinvestmentact/data/reference/geocorr2014.csv")
crosswalk.head()
```

```python
recovery_places_LIT = pd.merge(crosswalk, LMtracts, left_on='Tract', right_on='Tract') #code by place
#because this dataframe only has CRA eligible tracts, total loans is the sum of the loans to high income borrowers
#and low income borrowers in low income tracts only
recovery_places_LIT["Total Loans"] = recovery_places_LIT["# HI borrower, LI tract"]+ recovery_places["# LI borrower, LI tract"]

recovery_places_LIT.rename(columns={"# LI borrower, LI tract": "Loans to low-mod borrowers", 
                                                         "# HI borrower, LI tract": "Loans to mid-high borrowers"}, inplace=True)

recovery_places_LIT["CRA Eligible"] = recovery_places_LIT["CRA Eligible"].apply(lambda x: x.replace("not eligible", "0").replace("eligible", "1"))
recovery_places_LIT["CRA Eligible"].astype(int)
recovery_places_LIT.head()
```

```python
#recovery_places_LIT.groupby(["Place name"]).groups["Alameda city, CA"]

foo1 = recovery_places_LIT.groupby('Place name')['Total Loans'].sum()
foo2 = recovery_places_LIT.groupby('Place name')['Loans to mid-high borrowers'].sum()
```

```python
df_lol = []
for place, place_df in recovery_places_LIT.groupby("Place name"):
    pct_LCTs_HIB = place_df["# HI borrower, LI tract"]/place_df["Total Loans"]
    
    row_list = [place, place_df["County name"].iloc[0], place_df["Total Loans"].iloc[0], place_df["Loans to low-mod borrowers"].iloc[0],
                place_df["Loans to mid-high borrowers"].iloc[0], len(place_df["CRA Eligible"]), pct_LCTs_HIB.iloc[0]]
    df_lol.append(row_list)
                
                
data = pd.DataFrame(df_lol, columns=["Place Name", "County", "Total loans 2013-2017", "low-mod loans 2013-2017", 
    "mid-high loans 2013-2017", "num low-mod tracts", "pct loans to mid-high borrowers in low-mod tracts"])
data.head(40)   
```
